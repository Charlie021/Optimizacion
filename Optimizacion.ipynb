```python
import numpy as np
from matplotlib import pyplot
def f(x):
  return x/2-x*x
x = np.arange(0,1,0.01)
pyplot.plot(x, [f(i) for i in x])
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.xlim(0, 0.5)
pyplot.ylim(0, 0.07)
pyplot.title("Gráfica de Area Maxima")
pyplot.ylabel("Eje y")
pyplot.xlabel("Eje x")
pyplot.text(0.25,0.0625,"$ P (0.25,0.0625) $", color="b")
pyplot.plot(0.25,0.0625, 'r--o', label="Partícula 1")
pyplot.grid(ls="--", color="#dadada")
pyplot.savefig("función.png")
pyplot.show()
```
