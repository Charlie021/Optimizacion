**Ejercicio 1** \\
Un expendio de carnes prepara la carne para albondigon con una combinacion de carne molida de res y carne molida de cerdo. La carne de res contiene 80% carne y 20% grasa, y cuesta a la tienda 80 c. por libra. La carne de cerdo contiene 68% de carne y 32% de grasa, y cuesta 60 c. por libra.
¿Que cantidad de cada tipo de carne debe emplear la tienda en cada libra de albnondigon si se desea minimizar el costo y mantener la grasa menor al 25%? \\
**Variables de desicion:** \\
Carne de cerdo $x_1$ y carnde vaca $x_2$ \\
**Funcion Objetivo:** \\
Minimizar $60x_1+80x_2$ \\
**Restricciones:** \\
$0.32x_1+0.32x_2 \leq0.25$ \\
$x_1+x_2=1$ \\
$x_1,x_2\ge0$




```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure()
    
# Restricción para M1
plt.plot([0.25/0.32, 0], [0, 0.25/0.20], color='orange', alpha=0.25)
plt.fill_between([0.25/0.32, 0], [0, 0.25/0.20], color='orange', alpha=0.25, label='grasa 1')

plt.xlabel('Carne de cerdo')
plt.ylabel('Carne de vaca')
plt.legend()
plt.show()
```

```python
plt.figure()
    
# Restricción para M1
plt.plot([0.25/0.32, 0], [0, 0.25/0.20], color='orange', alpha=0.25)
plt.fill_between([0.25/0.32, 0], [0, 0.25/0.20], color='orange', alpha=0.25, label='Grasa 1')

# Restricción para M2
plt.plot([0, 1], [1, 0], color='blue', alpha=0.125)
plt.fill_between([0, 1], [1, 0], color='blue', alpha=0.125, label='Grasa 2')

plt.xlabel('recipientes K')
plt.ylabel('recipientes L')
plt.legend()
plt.show()
```
